**Условный опратор состоит из заголовка и тела оператора**:

````python
student_name = 'Вася'
price = 15000
if student_name == 'Вася':                          # заголовок условного оператора
    print('Васе положена скидка')                   # тело условного оператора
    price = price * 0.95
print(price)
````

Мы можем так же добавить дополнительные ветвления:

````python
student_name = 'Петя'
price = 15000
if student_name == 'Вася':                          # заголовок условного оператора
    print('Васе положена скидка')                   # тело условного оператора
    price = price * 0.95
else:                                               # else (еще) т.е. все другое или отличное от Вася
    print('Раз не Вася, то никакой скидки!')
print(price)
````

## 9. Функции

Функция - это часть кода, которая часто повторяется внутри программы. 

Например, представим себе, что мы хотим посчитать ежемесячный платеж по кредиту для нескольких вариантов. Формула расчета:

$ \large P = S\bigg( i + \frac{i} {(1+i)^n - 1} \bigg)$ 

где:
- $P$ – ежемесячный платёж по аннуитетному кредиту;
- $S$ – сумма кредита;
- $i$ – ежемесячная процентная ставка (годовая процентная ставка / 12);
- $n$ – срок, на который берётся кредит (количество месяцев).

Лирическое оступление. Посмотреть, как писать формулы с помощью latex-нотации можно вот <a href="https://ru.wikibooks.org/wiki/Математические_формулы_в_LaTeX">тут</a> и <a href="http://mydebianblog.blogspot.com/2009/01/latex-math-in-latex.html">вот тут</a>.

Напишем программу для расчета платежа:

````python
# 50000 рублей на 12 месяцев под 22% годовых
S = 50000
n = 12
annual = 0.22
i = annual / 12
P = S * (i + i/((1 + i) ** n - 1))
print(P)
````

````python
# 60000 рублей на 6 месяцев под 10% годовых
S = 60000
n = 6
annual = 0.1
i = annual / 12
P = S * (i + i/((1 + i) ** n - 1))
print(P)
````

````python
# 10000 рублей на 25 месяцев под 5% годовых
S = 10000
n = 25
annual = 0.05
i = annual / 12
P = S * (i + i/((1 + i) ** n - 1))
print(P)
````

Как видите, мы все время повторяем один и то же код. Хорошо было бы его где-то определить один раз и переиспользовать. Именно для этого существуют функции:

````python
def montly_payment(S, n, annual):              # заголовок функции; S, n, annual - входные параметры
    i = annual / 12                            # тело фунции
    P =  S * (i + i/((1 + i) ** n - 1)) 
    return round(P, 2)                         # результат функции
````

````python
montly_payment(50000, 12, 0.22) # 50000 рублей на 12 месяцев под 22% годовых
````

````python
montly_payment(60000, 6, 0.1) # 60000 рублей на 6 месяцев под 10% годовых
````

````python
montly_payment(10000, 25, 0.05) # 10000 рублей на 25 месяцев под 5% годовых
````

### Доп пример с факториалом числа (перемножение ряда чисел в промежутке от 1 до n)

````python
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result = result * i  # result *= i -  эквивалентно result * i,  а  result += 1 -  эквивалентно result = result + 1
    
    return result
````

````python
factorial(4)
````

````python
def factorial(n):
    if n <= 1:
        return 1
    
    return n * factorial(n - 1)
````

````python
factorial(4)
````
